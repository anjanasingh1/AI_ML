```python
# Import libraries
from sklearn.neighbors import KNeighborsClassifier

# Sample data
# Features: Temperature, Humidity, Wind
X = [
    [0, 0, 0],  # Hot, High, Weak
    [0, 0, 1],  # Hot, High, Strong
    [1, 0, 0],  # Mild, High, Weak
    [2, 1, 0],  # Cool, Normal, Weak
    [2, 1, 1],  # Cool, Normal, Strong
]

# Target: PlayTennis (No=0, Yes=1)
y = [0, 0, 1, 1, 0]

# Create KNN classifier
knn = KNeighborsClassifier(n_neighbors=3)

# Train the model
knn.fit(X, y)

# Predict for a new sample: Mild, Normal, Weak
sample = [[1, 1, 0]]
prediction = knn.predict(sample)

print("Prediction (1=Yes, 0=No):", prediction[0])
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

# Dataset with 10 samples
X = [
    [0, 0, 0],
    [0, 0, 1],
    [1, 0, 0],
    [1, 1, 0],
    [2, 1, 0],
    [2, 1, 1],
    [0, 1, 0],
    [1, 0, 1],
    [2, 0, 0],
    [1, 1, 1]
]
y = [0, 0, 1, 1, 1, 0, 0, 1, 1, 0]

# Test multiple k-values (odd numbers)
k_values = [1, 3, 5, 7]

print("Cross-validation results:")

for k in k_values:
    # KNN classifier
    knn = KNeighborsClassifier(n_neighbors=k)
    # 5-fold CV
    scores = cross_val_score(knn, X, y, cv=5)
    print(f"k={k}, Accuracy={np.mean(scores):.2f}")
```
