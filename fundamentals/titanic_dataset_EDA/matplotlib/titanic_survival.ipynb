```python
# importing required libraries
import numpy as np
import pandas as pd

# importing matplotlib
import matplotlib.pyplot as plt

# display plots in the notebook itself
%matplotlib inline
```

```python
data_BM = pd.read_csv('titanic.csv')
data_BM = data_BM.dropna(subset="boat")


data_BM.head()
```

```python
#Line Chart :With some datasets, you may want to understand changes in one variable as a function of time, or a similarly continuous variable.

groupBy_parch = data_BM.groupby('parch').survived.sum()

x = groupBy_parch.index.tolist()
y = groupBy_parch.values.tolist()

# set figure size
plt.figure(figsize=(14, 8))

# set title
plt.title('Line chart survived/parch')

# set axis labels
plt.xlabel('parch ')
plt.ylabel('survived')

# set xticks 
plt.xticks(labels=x, ticks=np.arange(len(x)))

plt.plot(x, y);
```

```python
# Bar Chart 
#A bar chart is another simple type of visualization that is used for categorical variables.
#You can use plt.bar() instead of plt.plot() to create a bar chart.
```

```python
groupBy_parch = data_BM.groupby('parch').survived.sum()

x = groupBy_parch.index.tolist()
y = groupBy_parch.values.tolist()

# set figure size
plt.figure(figsize=(14, 8))

# set title
plt.title('Bar chart survived/parch')

# set axis labels
plt.xlabel('parch ')
plt.ylabel('survived')

# set xticks 
plt.xticks(labels=x, ticks=np.arange(len(x)))

plt.bar(x, y, color=['red', 'orange', 'magenta','green','purple']);
```

```python
#Histograms are a very common type of plots when we are looking at data like height and weight, stock prices, waiting time for a customer, etc which are continuous in nature.
#Histogram’s data is plotted within a range against its frequency.

# set figure size
#plt.figure(figsize=(14, 8))

# set title
plt.title('hist chart survived/parch')

# set axis labels
plt.xlabel('survived ')
plt.ylabel('frequency')

# set xticks 
plt.xticks(labels=x, ticks=np.arange(len(x)))


# plot histogram
plt.hist(data_BM['survived'], bins=20, color='lightblue');
```

```python
#Box Plot helps to detect outliers
data = data_BM[['parch']]

# create outlier point shape
red_diamond = dict(markerfacecolor='r', marker='D')

# set title
plt.title('Box Plot / Parch/ Outliers')

# make the boxplot
plt.boxplot(data.values, labels=['parch'], flierprops=red_diamond);
```

```python
#Box Plot we can create subplots

data = data_BM[['parch', 'sibsp','survived']]

# create outlier point shape
red_diamond = dict(markerfacecolor='r', marker='D')

# generate subplots
fig, ax = plt.subplots()

# set title
plt.title('Box Plot - sub plots')

# make the boxplot
plt.boxplot(data.values, labels=['parch', 'sibsp','survived'], flierprops=red_diamond);
```

```python
#violin plot for density distribution

data = data_BM[['parch', 'sibsp']]

# generate subplots
fig, ax = plt.subplots()

# add labels to x axis
plt.xticks(ticks=[1,2], labels=['parch', 'sibsp'])

# make the violinplot
plt.violinplot(data.values);
```

```python
#scatter plot Relative distribution of item

# set label of axes 
plt.xlabel('parch')
plt.ylabel('survived')

# plot
plt.scatter(data_BM["parch"][:200], data_BM["survived"][:200]);
```

```python
#Bubble plot Relative distribution
# set label of axes 
plt.xlabel('parch')
plt.ylabel('sibsp')

# set title
plt.title('parch vs sibsp')

# plot
plt.scatter(data_BM["parch"][:100], data_BM["sibsp"][:100], s=data_BM["survived"][:100]*1000, c='red');
```
