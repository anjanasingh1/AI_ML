```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler

# Sample points
p1 = np.array([2, 3])
p2 = np.array([5, 7])

# 1. Euclidean Distance
euclidean = np.sqrt(np.sum((p1 - p2)**2))
print("Euclidean:", euclidean)

# 2. Manhattan Distance
manhattan = np.sum(np.abs(p1 - p2))
print("Manhattan:", manhattan)

# 3. Hamming Distance (categorical example)
a = np.array([1, 0, 1, 1])
b = np.array([1, 1, 0, 1])
hamming = np.sum(a != b)
print("Hamming:", hamming)

# -------------------------
# Scaling Example
data = np.array([[10], [20], [30], [100]])

# Min-Max Scaling
print("MinMax:", MinMaxScaler().fit_transform(data))

# Standard Scaling
print("Standard:", StandardScaler().fit_transform(data))

# Robust Scaling
print("Robust:", RobustScaler().fit_transform(data))
```
