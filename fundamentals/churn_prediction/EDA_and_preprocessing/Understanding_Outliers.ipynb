```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Simple data (one clear outlier: 100)
data = pd.DataFrame({'values': [10, 12, 14, 15, 18, 20, 22, 100,200]})
```

```python
sns.boxplot(y='values', data=data)
plt.title("Boxplot - Outlier Detection")
plt.show()
```

```python
Q1 = data['values'].quantile(0.25)
Q3 = data['values'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data[(data['values'] < lower_bound) | (data['values'] > upper_bound)]
print(outliers)
```

```python
sns.histplot(data['values'], bins=5)
plt.title("Histogram - Distribution")
plt.show()
```
